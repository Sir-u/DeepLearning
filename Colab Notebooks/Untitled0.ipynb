```python
import numpy as np

def every_step(step):
 return np.mean(step), np.argmin(step), np.min(step), np.argmax(step), np.max(step)
 

n = np.round(np.random.rand(30)*3000,0)
print(n)

step_mean, step_argmin, step_min, step_argmax, step_max = every_step(n)

print(step_mean, step_argmin+1, step_min, step_argmax+1, step_max)
```
